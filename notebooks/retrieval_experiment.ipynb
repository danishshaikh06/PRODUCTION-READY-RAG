```python
"""
Diagnose why metadata_search on recipient_names returns 0 results
despite the data existing in Postgres.
"""

from qdrant_client import QdrantClient
from my_rag_app.constants import QDRANT_URL, QDRANT_COLLECTION
from my_rag_app.entity.models import Metadata
from my_rag_app.config.config import get_session

client = QdrantClient(url=QDRANT_URL)

# 1. Check Postgres directly — does this name actually exist in the metadata table?
print("--- Postgres check ---")
with get_session() as session:
    rows = session.query(Metadata).filter(
        Metadata.recipient_names.any("Quang Trung Nguyễn")
    ).all()
    print(f"Postgres rows with exact match: {len(rows)}")

    # Broader check — any row containing "Trung" in any recipient name
    all_rows = session.query(Metadata).filter(Metadata.recipient_names != []).limit(500).all()
    matches = [r for r in all_rows if any("Trung" in n for n in r.recipient_names)]
    print(f"Rows where any recipient_name contains 'Trung': {len(matches)}")
    for r in matches[:5]:
        print("  email_id:", r.email_id, "| recipient_names:", r.recipient_names)

print()

# 2. Check Qdrant payload directly — what does recipient_names actually look like in a real point?
print("--- Qdrant payload check ---")
points, _ = client.scroll(collection_name=QDRANT_COLLECTION, limit=5, with_payload=True)
for p in points:
    print("recipient_names in payload:", p.payload.get("recipient_names"))

print()

# 3. Scroll Qdrant for ANY point where recipient_names is non-empty, then check for Trung manually
print("--- Manual scan in Qdrant for 'Trung' ---")
all_points, _ = client.scroll(collection_name=QDRANT_COLLECTION, limit=1000, with_payload=True)
found = [p for p in all_points if any("Trung" in n for n in p.payload.get("recipient_names", []))]
print(f"Points where recipient_names contains 'Trung' (manual scan): {len(found)}")
for p in found[:3]:
    print("  recipient_names:", p.payload.get("recipient_names"))
```
