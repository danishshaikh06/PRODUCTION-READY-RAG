```python
retriever = HybridRetriever(qdrant_url=QDRANT_URL, collection_name=QDRANT_COLLECTION)
results = retriever.search("NOC from MoCA to operate cargo flight at VAOZ airport", top_k=10)
reranker = CrossEncoderReranker()
top_results = reranker.rerank("NOC from MoCA to operate cargo flight at VAOZ airport", results, top_k=3)

for r in top_results:
    print(f"{r['score']:.4f} | {r['payload']['text'][:500]}")
```
